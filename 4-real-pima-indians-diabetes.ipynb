# Real pima indians diabetes
## Setup

```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

## Make sure we get the same value

This is because we're using a stochastic model, it means we"ll get the same answer every time because we're essentially disabling randomness for a little bit.

```python
seed = 7
np.random.seed(seed)
```

## Lets have a sneaky peak at that data

```python
df = pd.read_csv("diabetes.csv")
df
```

```python
x = df.iloc[:, 0:8].to_numpy()
y = df.iloc[:, 8].to_numpy()
x
```

```python
y
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.33, random_state=seed)
```

## Now the keras stuff

```python
model = tf.keras.Sequential()
model.add(
    tf.keras.layers.Dense(
        12,
        kernel_initializer=tf.keras.initializers.random_uniform,
        activation=tf.keras.activations.relu
    )
)
model.add(
    tf.keras.layers.Dense(
        8,
        kernel_initializer=tf.keras.initializers.random_uniform,
        activation=tf.keras.activations.relu
    )
)
model.add(
    tf.keras.layers.Dense(
        8,
        kernel_initializer=tf.keras.initializers.random_uniform,
        activation=tf.keras.activations.relu
    )
)
model.add(
    tf.keras.layers.Dense(
        1,
        kernel_initializer=tf.keras.initializers.random_uniform,
        activation=tf.keras.activations.sigmoid
    )
)

model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=[tf.keras.metrics.binary_accuracy]
)
```

## Fitting

```python
model.fit(train_x, train_y, epochs=500, batch_size=10)
```

## Now we see how well it worked

```python
scores = model.evaluate(test_x, test_y)
scores
```

Hmm, that's not super helpful

```python
print(f"{model.metrics_names[0]}: {scores[0] * 100}")
print(f"{model.metrics_names[1]}: {scores[1] * 100}%")
```

## Lets use it!

```python
preds = model.predict(
    pd.DataFrame(
        [{
            "Pregnancies": 6, "Glucose": 148, "BloodPressure": 72, "SkinThickness": 35,
            "Insulin": 169.5, "BMI": 33.6, "DiabetesPedigreeFunction": 0.627,
            'Age': 50
        }]
    ).to_numpy()
)

preds
```
